## Изменение структуры доходов бюджета Санкт-Петербурга (2016—2024)

Источник данных: https://www.iminfin.ru/areas-of-analysis/budget/finansoviy-pasport-subjecta-rf/dokhody-detalno?territory=1

```python
import pandas as pd
```

```python
data_2024 = pd.read_csv('income_2024.csv', encoding='1251')
data_2023 = pd.read_csv('income_2023.csv', encoding='1251')
data_2022 = pd.read_csv('income_2022.csv', encoding='1251')
data_2021 = pd.read_csv('income_2021.csv', encoding='1251')
data_2020 = pd.read_csv('income_2020.csv', encoding='1251')
data_2019 = pd.read_csv('income_2019.csv', encoding='1251')
data_2018 = pd.read_csv('income_2018.csv', encoding='1251')
data_2017 = pd.read_csv('income_2017.csv', encoding='1251')
data_2016 = pd.read_csv('income_2016.csv', encoding='1251')
```

```python
data_2024 = data_2024[:17]
data_2023 = data_2023[:17]
data_2022 = data_2022[:17]
data_2021 = data_2021[:17]
data_2020 = data_2020[:17]
data_2019 = data_2019[:17]
data_2018 = data_2018[:17]
data_2017 = data_2017[:17]
data_2016 = data_2016[:17]
```

```python
describe = data_2022.loc[:]['Показатели']
```

```python
describe
```

```python
aim = 'Уточненный план/конс.бюджет субъекта РФ'
aim_2024 = data_2024.loc[:][aim]
aim_2023 = data_2023.loc[:][aim]
aim_2022 = data_2022.loc[:][aim]
aim_2021 = data_2021.loc[:][aim]
aim_2020 = data_2020.loc[:][aim]
aim_2019 = data_2019.loc[:][aim]
aim_2018 = data_2018.loc[:][aim]
aim_2017 = data_2017.loc[:][aim]
aim_2016 = data_2016.loc[:][aim]
```

```python
def to_int(ser):
    for i in range(len(ser)):
        ser[i] = ser[i].split()
        ser[i] = ''.join(ser[i])
        ser[i] = ser[i].split(',')[0]
        ser[i] = int(ser[i])
    return ser
```

```python
aim_2024 = to_int(aim_2024)
aim_2023 = to_int(aim_2023)
aim_2022 = to_int(aim_2022)
aim_2021 = to_int(aim_2021)
aim_2020 = to_int(aim_2020)
aim_2019 = to_int(aim_2019)
aim_2018 = to_int(aim_2018)
aim_2017 = to_int(aim_2017)
aim_2016 = to_int(aim_2016)
```

```python
aim_2017
```

```python
aim_2023
```

```python
dynamic = pd.DataFrame(data={'Показатели': describe,
                             '2016': aim_2016,
                             '2017': aim_2017,
                             '2018': aim_2018,
                             '2019': aim_2019,
                             '2020': aim_2020,
                             '2021': aim_2021,
                             '2022': aim_2022,
                             '2023': aim_2023,
                             '2024': aim_2024,
                            })
```

```python
dynamic = dynamic.T
```

```python
dynamic.columns = dynamic.loc['Показатели']
dynamic = dynamic.drop('Показатели')
```
